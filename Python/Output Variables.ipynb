```python
#Output variables: The Python print() function is often used to output variables.
x = "Python is awesome"
print(x)
```

```python
#In the print() function, you output multiple variables, separated by a comma:
x = "Python"
y = "is"
z = "awesome"
print(x, y, z)
```

```python
#Use of + operator to make an statement from multiple variables:
x = "Python"
y = "is"
z = "awesome"
print(x+' '+ y+ z)
print(x+' '+y+' '+z)
```

```python
#The best way to output multiple variables in the print() function is to separate them with commas, which even support different data types: as we can't add int with string
x=100
y='John'
#print(x+y) [unsupported operand type(s) for+: 'int' & 'str']
print(x,y)
```
